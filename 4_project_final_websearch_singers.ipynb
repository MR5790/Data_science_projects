```python
class relatedSearchanlsys():
    def __init__(self,relatedSearch1,dfnew):
        self.words = relatedSearch1['words']
        self.index = relatedSearch1['index']
        self.Artist=relatedSearch1['Artist'] 
        self.totalCount=dfnew['totalCount']
    #search the number related words abut each artist and the unpopular word search on top of the list
    def grouping_words_count(self,relatedSearch1):
        n1=relatedSearch1.groupby(['Artist','index'], as_index=False)['words'].count() 
        n2=n1.sort_values(by=['words'])
        return n2
    #artists with the least number of searches on the web on the bottom of the list
    def most_popular_artists(self,dfnew):
        n3=dfnew.sort_values(by=['totalCount'])
        return n3
    
```

```python
relatedSearchanlsysnew=relatedSearchanlsys(relatedSearch1,dfnew)
```
